##7. Detección de datos atipicos (Outliers)
- Aqui utilizaremos los graficos de box (caja) para identificar los datos atipicos de cada variable, sus medias y comparaciones.

```python
sns.boxplot(x=df['Recency'])
plt.title('Número de días desde la última compra del cliente')
plt.show()
```

```python
df['Recency'].describe()
```

 ¿Las distribuciones son simétricas o están sesgadas hacia algún lado? (hablando de la mediana)

R: Segun el grafico de caja, podemos observar que en la linea central o su mediana no tiende a ser sesgada por lo que cliente compra entre mas 49 dias o menos de los 49 dias ya que la media apunta al 50%.


```python
#---- Columnas ---#
# ID	Year_Birth	Education	Marital_Status	Income
# Kidhome	Teenhome	Dt_Customer	Recency	MntWines	...
# NumWebVisitsMonth	AcceptedCmp3	AcceptedCmp4	AcceptedCmp5
# AcceptedCmp1	AcceptedCmp2	Complain	Z_CostContact	Z_Revenue	Response
sns.boxplot(x=df['Income'])
plt.title(' Ingresos familiares anuales del cliente')
plt.show()
```

```python
df['Recency'].describe()
```

¿Algún grupo tiene una mayor dispersión? (si la caja es mas ancha ) si tiene mayor IQR rango intercuartilico significa que tiene mayor dispersion

R: Podemos encontrar varios valores atipicos en el grafico de caja. Tambien decir que hay un dato atipico que excede el valor de Ingresos familiares anuales del clientes la cual seria 900000.

R: La caja representa el rango intercuartílico (IQR), que abarca el 50% central de los datos. Cuanto mayor sea el IQR , mayor es la dispersión en los datos. En este caso, la dispersion de los datos es super reducida ya que el grafico de caja no es ancha sino angosta.


##8. Histogramas, Graficos de dispersion y Grafico de barras


- Analisaremos en resumen y encotraremos algun patron entre los diferentes graficos de dispersion y histogramas.

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['Income'])
plt.title('Distribución de Ingresos Familiares Anuales')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.show()
```

```python
df['Income'].median() #Mediana del histograma
```

- podemos notar que en el histograma de la variable ingresos podemos notar una alta dispersion de los datos centrados.

```python

plt.figure(figsize=(8, 6))
plt.bar(df['Education'].unique(), df['Education'].value_counts())
plt.xlabel("Nivel de Educación")
plt.ylabel("Cantidad de Clientes")
plt.title("Distribución de Clientes por Nivel de Educación")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

- Podemos notar que mas de 1000 clientes lograron el nivel de educacion de graduacion

- Mientras que un pequeño porcentaje que solo quedaron en el 2do ciclo
